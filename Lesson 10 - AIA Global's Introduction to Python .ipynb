## COVID-19 Example of using Python to automate taking and visualizing data

# dependencies

```python
import pandas as pd
import datetime as dt

import matplotlib.dates as mdates
from matplotlib import pyplot as plt
from matplotlib import rcParams
```

# settings

```python
pd.options.display.max_rows = 500        #Increases the number of rows displayed without a break in the middle.
pd.options.display.max_columns = 100     #Increases the number of columns displayed without a break in the middle.
pd.set_option('max_colwidth',100)        #Increases the width of the columns to display 100 characters at the same time.
```

# data location

```python
url_deaths    = 'https://github.com/CSSEGISandData/COVID-19/blob/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'
```

# read the data into a dataframe

```python
# read the datainto a dataframe
deaths    = pd.read_html(url_deaths)[0]


# drop the first column. 
deaths    = deaths.drop(columns = 'Unnamed: 0')
```

# display the data

```python
print('\nDeaths:')
display(deaths.head(20))
display(deaths.tail(20))
```

# functions and definitions

```python
def plot_country_data(df, country, figure_title):
    plt.style.use('seaborn-whitegrid')
    country_subset = df.loc[df['Country/Region']==country].copy()
    country_subset = country_subset.drop(columns = ['Lat','Long'])
    country_subset.loc['Total'] = country_subset.sum(numeric_only=True, axis=0)
    country_subset = country_subset.fillna('')

    total_subset = (country_subset.iloc[-1:,]).copy()
    total_subset = total_subset.iloc[:,2:]

    difference_subset = total_subset.copy()
    difference_subset = difference_subset.diff(axis = 1, periods = 1).rename(index = {'Total':'Daily'})

    dates = list(total_subset.columns)
    dates = [dt.datetime.strptime(d,'%m/%d/%y').date() for d in dates]

    x1 = dates
    y1 = list(total_subset.iloc[0].values)

    x2 = dates
    y2 = list(difference_subset.iloc[0].values)

    # plot intitial text
    source_text_1 = 'DATA SOURCE:  2019 Novel Coronavirus COVID-19 (2019-nCoV) Data Repository by Johns Hopkins CSSE'
    
    rcParams['figure.figsize'] = 18, 1
    plt.figtext(0,0.4,figure_title,fontsize = 24,fontweight='bold')
    plt.figtext(0,0,source_text_1, fontsize = 18)
    plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    # plot graphs and layouts

    rcParams['figure.figsize'] = 18, 12
    fig = plt.figure()
    ax1 = plt.subplot2grid((6, 4), (0, 0), rowspan=2, colspan=4)
    ax2 = plt.subplot2grid((6, 4), (2, 0), rowspan=2, colspan=4)
    ax3 = plt.subplot2grid((6, 4), (4, 0), rowspan=2, colspan=4)

    ax1.plot(x2,y2, 'black', linestyle = '--')
    ax1.bar(x2,y2,color = 'salmon', alpha = 0.7)
    ax1.set_title('Daily Count', fontname='DejaVu Sans',fontweight = 'bold',fontsize = 18)
    ax1.set_xlabel('Date', fontsize = 14)
    ax1.set_ylabel('Daily Count', fontsize = 14)
    ax1.grid(b = True, which = 'both', color = 'black',linestyle = '-', alpha = 0.4, linewidth=1)

    ax2.plot(x1,y1, color = 'black', linestyle = '--')
    ax2.bar(x1,y1,color = 'skyblue', alpha = 0.7)
    ax2.set_title('Total Count Over Time', fontname='DejaVu Sans',fontweight = 'bold', fontsize = 18)
    ax2.set_xlabel('Date', fontsize = 14)
    ax2.set_ylabel('Total Count', fontsize = 14)
    ax2.grid(b = True, which = 'both', color = 'black',linestyle = '-', alpha = 0.4, linewidth=1)

    ax3.plot(x1,y1, color = 'black', linestyle = '--')
    ax3.bar(x1,y1,color = 'orange', alpha = 0.7)
    ax3.set_title('Total Count Over Time on a Log Scale',fontname='DejaVu Sans',fontweight = 'bold', fontsize = 18)
    ax3.set_xlabel('Date', fontsize = 14)
    ax3.set_ylabel('(Log Scale)', fontsize = 14)
    ax3.set_yscale('log')
    ax3.grid(b = True, which = 'both', color = 'black',linestyle = '-', alpha = 0.4, linewidth=1)

    plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
    plt.gca().xaxis.set_major_locator(mdates.DayLocator(interval=5))
    plt.tight_layout()
    plt.show()
    
```

```python
plot_country_data(deaths, 'US', figure_title = 'US Deaths Through 3/27/2020' )
```
